# Students' Marks Analysis Project

Description: This project analyzes marks of 50 students across 3 subjects using NumPy.
Includes total, average, highest, pass/fail stats, and more.


```python
import numpy as np
```

```python
students_marks = np.random.randint(101,size = (50,3))

total_marks=np.sum(students_marks, axis=1)

print("Total marks of each student:")
for i in total_marks:
    print(i)
    
```

```python
print("Average marks of each student:")
for i in total_marks:
    avg_mark = round(i/3,1)
    print(avg_mark)
```

```python
print("Highest score of each student:")
for row in students_marks: #accessing rows
    highest_mark = np.max(row) 
    print(highest_mark)
```

```python
print("Lowest score of each student:")
for row in students_marks:
    lowest_mark = np.min(row)
    print(lowest_mark)
```

```python
print("average number in each subject:")
sum_of_subj = np.sum(students_marks, axis = 0)
avg_of_subj = sum_of_subj/50
print(avg_of_subj)
```

```python
print("Maximum marks in each subject:")
for i in students_marks.T: #iterating thru each column of original array thru transpose
    print(max(i))

print("Minimum marks in each subject")
for i in students_marks.T: 
    print(min(i))
```

```python
print("No. of students with more than 90 marks:")
above90 = np.sum(np.any(students_marks>90,axis=1))
print(above90)
```

```python
print("No. of students who passed all exams:")
passing_marks = 35
passed_stud = np.sum(np.all(students_marks>35,axis=1)) 
print(passed_stud)

print("No. of students who failed any exam:")
failed_stud = np.sum(np.any(students_marks<passing_marks,axis=1)) 
print(failed_stud)
```

```python
print("TOP 5 students of the class:")
sorted_total=np.sort(total_marks)
top5=sorted_total[-1:-6:-1]
print(top5)

print("Bottom 5 students of the class:")
bottom5 = sorted_total[:5]
print(bottom5)
```

```python
print("No. of students scored less than 20 in maths(1st column)")
maths_score = students_marks[:,0]
below20 = np.sum(maths_score>20)
print(below20)
```

```python
#SUMMARY
print("Student who topped the class has a total score:", max(total_marks))
print("The top students are:", top5)
print("Each subject averages:", avg_of_subj)
print("No. of students who passed all exams:",passed_stud)
print("No. of students who passed any exams:",failed_stud)
```
